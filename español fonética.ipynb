## 六 强元音、弱元音、二重元音
**1. 强元音(发音强一点，重一点，重读)**  
>    a e o  

**2. 弱元音(发音若一点，轻一点，轻读)**  
>    i u  

**3. 二重元音**  

    两个元音字母组合，其中至少有一个弱元音（重音放在强元音上） 
    ia ie io ua ue uo  
    ai ei oi au eu ou  
    iu ui  

    güe güi					u加两点后要发音，组合后变为二重元音  
    gue gui					u不发音  
## 七 辅音连缀
    当l和r放在p,b,t,d,c,g,f后面，组成辅音连缀
    
>  pl pr bl br  
   tl tr dl dr  
   cl cr gl gr fl fr  
    
    pl,pr,bl,br
>  pla ple pli plo plu  
   pra pre pri pro pru  
   bla ble bli blo blu  
   bra bre bri bro bru  
   
    tl,tr,dr
>  tla tle tli tlo tlu  
   tra tre tri tro tru  
   dra dre dri dro dru  
   
    cl,cr,gl,gr
>  cla cle cli clo clu  
   cra cre cri cro cru  
   gla gle gli glo glu  
   gra gre gri gro gru  
   
    fl,fr
>  fla fle fli flo flu  
   fra fre fri fro fru  

## 八 音节
**1. 音节**  

    音节就是发音的最小单元
**2. 音节的核心**  

    音节核心就是元音，划分音节就是找元音  
**3. 划分音节**  

>  Ema E-ma    Ana A-na  
   Susana Su-sa-na    Manolo Ma-no-lo  
   palma pal-ma    asma as-ma 

    二重元音是一个音节，要划在一起
>  piano pia-no    bueno bue-no  

    辅音连缀也要划在一个音节内
>  aplica a-pli-ca    Pedro Pe-dro  

    二重元音和辅音连缀出现在一个音节中
>  aplauso a-plau-so  
   flauta flau-ta  

    注意的情况
>  televisión te-le-vi-sión  
   dia dí-a fílsosofia fí-lo-so-fí-a  
   feo fe-o tarea ta-re-a  

## 九 重音规则
**1. 如果单词以元音(a,e,i,o,u)或者以辅音(s,n)结尾，单词的重音放在倒数第二个音节上**  
>mano m**a**-no    Susana Su-s**a**-na  
mesas m**e**-sas    piano pi**a**-no  
tienen ti**e**-nen    libros l**i**-bros  

**2. 如果以其它辅音结尾(s,n除外)，单词的重音放在倒数第一个音节**  
>papel pa-p**e**l    feliz fe-l**i**z  
tener te-n**e**r    entrad en-tr**a**d  
Miguel Mi-gu**e**l    Samuel Sa-mu**e**l  

**3. 如果单词上有重音符号，则在重音符号上重读**  
>lápiz l**á**-piz    pésimo p**é**-si-mo  
filosofía f**i**-lo-so-f**í**-a  
útil **ú**-til    anímate a-n**í**-ma-te
